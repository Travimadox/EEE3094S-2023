```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from the CSV file
file_path = 'ControllerData/HelicopterData1.csv' 
data = pd.read_csv(file_path)

# Calculate values for horizontal lines
set_point = 8.884
upper_limit = set_point * 1.05  # +5%
lower_limit = set_point * 0.95  # -5%

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(data['Time(s)'], data['Output(m)'], label='Output (m)', color='b')
plt.axhline(y=set_point, color='r', linestyle='--', label='Set Point')
plt.axhline(y=upper_limit, color='g', linestyle='--', label=f'+5% Line at {upper_limit:.3f} m')
plt.axhline(y=lower_limit, color='g', linestyle='--', label=f'-5% Line at {lower_limit:.3f} m')
plt.xlabel('Time (s)')
plt.ylabel('Output (m)')
plt.title('Output in Meters vs Time')
plt.legend()
plt.grid(True)
plt.show()

# Find maximum output and calculate overshoot
max_output = data['Output(m)'].max()
overshoot = ((max_output - set_point) / set_point) * 100

# Find final output and calculate steady-state error
final_output = data['Output(m)'].iloc[-1]
steady_state_error = ((set_point - final_output) / set_point) * 100

print(f"Maximum Output: {max_output:.3f} m")
print(f"Overshoot: {overshoot:.2f}%")6.6
print(f"Final Output: {final_output:.3f} m")
print(f"Steady-State Error: {steady_state_error:.2f}%")
```
